```python
combined_df = background_treatment_epinephrine_df.merge(procedure_df, on='PcrKey', how='left')
combined_df



```

```python
columns_to_drop = ['Gender_Not Applicable', 'Gender_Not Recorded', 'Gender_Unknown (Unable to Determine)', 
                   'Race_Not Applicable', 'Race_Not Recorded']
combined_df.drop(columns = columns_to_drop, inplace=True)
combined_df.head()

```

```python
# Fill nan values for medical treatment and procedure columns
combined_df[new_columns_to_impute] = combined_df[new_columns_to_impute].fillna(0)



```

```python
combined_df[combined_df['PcrKey'] == 131432252][['PcrKey', 'EpinephrineUsed', 
                                                 'EpinephrineFrequency', 
                                                 'EpinephrineTotalAmount']]


```

```python
from sklearn.impute import SimpleImputer

# Impute one-hot encoded features with mode
mode_imputer = SimpleImputer(strategy='most_frequent')
combined_df[new_columns_to_impute] = mode_imputer.fit_transform(combined_df[new_columns_to_impute])

```
