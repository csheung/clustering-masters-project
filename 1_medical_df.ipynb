```python
# Drop answers of "Not Applicable" and "Not Recorded"
filtered_med_df = med_df[~((med_df.eMedications_03 == 7701001) | (med_df.eMedications_03 == 7701003))]

```

```python
def create_simple_one_hot_col(df, col, new_col, value_lists):
    """
    function: turn certain values in an existing column into a new one-hot-encoded column.
    return: nothing, modify the passed-in DataFrame in-place.
    """
    df[new_col] = df[col].isin(value_lists).astype(int)
    
```

```python
# Step 1: Identify if Epinephrine was used
create_simple_one_hot_col(df = filtered_med_df, 
                          col = 'eMedications_03', 
                          new_col = 'EpinephrineUsed', 
                          value_lists = [317361, 328316])
```

```python
# Review the newly created column of EpinephrineUsed
filtered_med_df[:]
```

```python
# locate the row indexes of selected values
def loc_row_indexes_in_single_col(df, col, value_lists):
    """
    function: locate the row indexes of selected values
    return: a list of Bool, indicating if the value exists
    """
    return df[col].isin(value_lists)

# Step 1: Identify rows where Epinephrine was used
mask = loc_row_indexes_in_single_col(filtered_med_df, 'eMedications_03', [317361, 328316])
mask[:]
```

```python
# Create a new column in the df to map the value based on an existing column
def map_values_to_new_col(df, index_mask, col, new_col, value_map):
    """
    function: Create a new column in the df to map the value based on an existing column
    return: nothing, new column added in-place
    """
    df.loc[index_mask, new_col] = df.loc[index_mask, col].map(value_map)
    df[new_col].fillna(0, inplace=True) # fill NaN with 0 <- not in map, no Epinephrine used
    
```

```python
# Step 2: Calculate the dosage based on the code
map_values_to_new_col(df = filtered_med_df,
                      index_mask = mask,
                      col = 'eMedications_03',
                      new_col = 'EpinephrineDosage',
                      value_map = {317361: 0.1, 328316: 1})

filtered_med_df[:]
```

```python
# Calculate the total amount of Epinephrine
def calculate_sum_group_by_pcr_key(df, key, col, new_col):
    """
    function: Calculate the Epinephrine Dosage Total Amount, 
            put them in a new column grouped by primary key
    return: new DataFrame with the new column showing the total amount
    """
    new_df = df.groupby(key)[col].sum().reset_index()
    new_df.rename(columns = {col : new_col}, inplace=True)
    return new_df

```

```python
# Step 3a: Calculate the frequency of Epinephrine used for each PcrKey -> group by PcrKey
frequency_med_df = calculate_sum_group_by_pcr_key(filtered_med_df, 'PcrKey', 'EpinephrineUsed', 'EpinephrineFrequency')
frequency_med_df

```

```python
# Step 3b: Calculate the total amount of Epinephrine used for each PcrKey -> group by PcrKey
total_dosage_med_df = calculate_sum_group_by_pcr_key(filtered_med_df, 'PcrKey', 'EpinephrineDosage', 'EpinephrineTotalAmount')
total_dosage_med_df.head()
```

```python
# Only keep one row of each duplicated PcrKey after calculating Epinephrine amount
# [!!] XOR operation on the EpinephrineUsed column [!!]
filtered_med_df2 = filtered_med_df.drop(columns=["eMedications_03", "eMedications_03Descr", "EpinephrineDosage"])  # Drop prev used columns

# Aggregate 'EpinephrineUsed' column using logical OR (any '1' remains '1')
agg_functions = {'EpinephrineUsed': 'max'}
# Apply aggregation and then drop duplicates
med_df_unique = filtered_med_df2.groupby('PcrKey').agg(agg_functions).reset_index()

med_df_unique[:]

```

```python
# Step 4a: Merge the frequency back to the original dataframe
grouped_med_df1 = pd.merge(med_df_unique, frequency_med_df, on='PcrKey', how='left')

```

```python
# Step 4b: Merge the total dosage back to the original dataframe
grouped_med_df2 = pd.merge(grouped_med_df1, total_dosage_med_df, on='PcrKey', how='left')


```
