```python
def create_rename_dict(filepath, df, int_type_convert_list, index, column):
    # build dict for column labels
    df = pd.read_csv(filepath)
    
    # convert type to int to get rid of decimals
    if int_type_convert_list:
        hf.convert_type_to_int(df, int_type_convert_list)
        
    # Convert the DataFrame to a dictionary for renaming
    rename_dict = df.set_index(index)[column].to_dict()
    
    return rename_dict
```

```python
def capitalize_labels_in_rename_dict(dictionary):
    for key, val in dictionary.items():
        dictionary[key] = ''.join(word.title() for word in val.split())
```

```python
# Create dictionary to translate the code to understandable col names
procedure_rename_dict = create_rename_dict(filepath = 'processed_data_cardiac_arrest/eProcedures_dict_airway.csv',
                                           df = procedure_df,
                                           int_type_convert_list = ["SNOMED"],
                                           index = "SNOMED",
                                           column = 'EMS_DESCRIPTION')
```

```python
# Convert all the dict key as strings for easier operations
def convert_dict_key_to_str(prev_dict):    
    # add new string-typed key into dict
    new_dict = {}
    for k, v in prev_dict.items():
        new_dict[str(k)] = v
    return new_dict
        
procedure_rename_dict = convert_dict_key_to_str(procedure_rename_dict)
    
```

```python
# Check if all keys in the dict are of the same data type
set(type(k) for k in procedure_rename_dict.keys())

```

```python
# filter data relevant to Airway treatment
filtered_procedure_df = procedure_df[procedure_df['eProcedures_03'].isin(procedure_rename_dict)]

```

```python
def create_one_hot_encoded_table(df, column_name, rename_dict):

    # One-hot encode the 'eProcedures_03' Airway column
    one_hot_encoded = pd.get_dummies(df[column_name]).astype(int)

    # Rename columns using the provided dictionary
    return one_hot_encoded.rename(columns=rename_dict)
    
def concat_one_hot_encoded_table(df, column_name, rename_dict):
    # Create the one-hot encoded table
    one_hot_encoded = create_one_hot_encoded_table(df, column_name, rename_dict)
    
    # Merge the one-hot encoded columns with the original DataFrame
    combined_df = pd.concat([df, one_hot_encoded], axis=1)
    
    # drop the original column
    combined_df.drop(columns=[column_name], inplace=True)
    
    return combined_df

```

```python
procedure_one_hot_encoded = create_one_hot_encoded_table(filtered_procedure_df, 'eProcedures_03', procedure_rename_dict)
procedure_one_hot_encoded.head()

```

```python
procedure_one_hot_encoded.to_csv('clustering_data/demo_procedure_airway_cols_no_pcr_key.csv', index=False)

```

```python
# Merge the one-hot encoded columns with the original DataFrame
proc_combined_df = concat_one_hot_encoded_table(filtered_procedure_df, 'eProcedures_03', procedure_rename_dict)
proc_combined_df.head()

```

```python
proc_combined_df.to_csv('clustering_data/demo_procedure_airway_cols_with_pcr_key.csv', index=False)

```

```python
def group_pcr_key(df, key_column):
    # Group by 'PcrKey' and sum up the one-hot encoded values for each key
    return df.groupby(key_column).max().reset_index()
                  

```

```python
procedure_df_with_unique_pcr_keys = group_pcr_key(proc_combined_df, 'PcrKey')
procedure_df_with_unique_pcr_keys.head()


```
